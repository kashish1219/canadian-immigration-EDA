```python
politics_in_canada = pd.DataFrame(
    {
        'imm_total':total,
        'Ruling_Party':pd.Series(ruling_party)
    }
)
politics_in_canada
```

```python
custom = {'Liberals': 'red','Conservative':'blue'}
plt.figure(figsize=(16,8))
ax1 = sns.scatterplot(
                x=politics_in_canada.index,
                y=politics_in_canada['imm_total'], 
                hue=politics_in_canada.Ruling_Party,
                palette=custom,
                s=100)
plt.xlabel('Years')
plt.ylabel('Immigration')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
hs = pd.read_csv('houses.csv')
hs.head(5)
```

```python
hs = hs.set_index('Unnamed: 0')
hs = hs.loc[['Housing estimates','Housing completions']]
```

```python
hs
```

```python
hs.columns = hs.iloc[0]
hs = hs.drop('Housing estimates', axis=0)
```

```python
hs = hs.iloc[:, :-2]
```

```python
hs = hs.loc['Housing completions'].reset_index()
hs = hs.iloc[1:]
hs = hs.set_index('Housing estimates')
hs['Housing completions'] = hs['Housing completions'].replace({',': ''}, regex=True).astype(int)
```

```python
plt.figure(figsize=(16, 8))

ax1 = sns.scatterplot(
                x=politics_in_canada.index,
                y=politics_in_canada['imm_total'], 
                hue=politics_in_canada.Ruling_Party,
                palette=custom,
                s=100)
plt.plot(hs.index, hs['Housing completions'], color='green', linewidth=2.5, label='Houses Built')
plt.legend()
plt.xlabel('Years')
plt.ylabel('Immigration')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
politics_in_canada=politics_in_canada[:-1]
df = politics_in_canada.join(hs)
df
```

```python
party_stats = df.groupby('Ruling_Party')[['imm_total', 'Housing completions']].mean().round(2)
party_stats['imm_per_house'] = (party_stats['imm_total']/party_stats['Housing completions'])
party_stats
```

On average, during Liberal governments, about 1.37 immigrants arrived per house built, compared to 1.11 under Conservatives. While this suggests a slightly higher housing pressure under Liberal governments, the ratio doesn't account for broader housing demand or supply-side dynamics â€” including domestic needs, emigration, or housing delays.

### Plotting

```python
labels = ['Liberals', 'Conservative']
ticks = np.arange(len(labels))
imm_vals = [party_stats.loc['Liberals', 'imm_total'], party_stats.loc['Conservative', 'imm_total']]
houses = [party_stats.loc['Liberals', 'Housing completions'],party_stats.loc['Liberals', 'Housing completions']]
width = 0.35
fig,ax = plt.subplots(figsize=(10,6))
bar1 = ax.bar(ticks-width/2, imm_vals, width, label='immigration', color='lightgreen')
bar2 = ax.bar(ticks+width/2, houses, width, label='Houses built', color='lightgrey')

ax.set_ylabel('Average Count')
ax.set_title('Average Immigration vs Housing Completions by Party')
ax.set_xticks(ticks)
ax.set_xticklabels(labels)
ax.grid(axis='y', linestyle='--', alpha=0.5)

for bar_group in [bar1, bar2]:
    for bar in bar_group:
        y = bar.get_height()
        ax.text(bar.get_x()+width/2,y+2000, f'{y:.0f}', ha='center', va='bottom', fontsize=9)
        
plt.tight_layout()
plt.show()
```

## Descriptive Statistics
We begin with basic descriptive statistics to understand the scale, spread, and central tendencies in immigration and housing data.

```python
df[['imm_total', 'Housing completions']].describe()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['imm_total'], kde=True, color='skyblue')
plt.title('Distribution of Immigration')

plt.subplot(1, 2, 2)
sns.histplot(df['Housing completions'], kde=True, color='salmon')
plt.title('Distribution of Housing Completions')

plt.tight_layout()
plt.show()
```

## Trends Over Time
Let's explore how immigration and housing completions evolve over time using percentage changes and rolling averages.

```python
df['imm_pct_change'] = df['imm_total'].pct_change() * 100
df['house_pct_change'] = df['Housing completions'].pct_change() * 100

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['imm_pct_change'], label='Immigration % Change', color='green')
plt.plot(df.index, df['house_pct_change'], label='Housing % Change', color='orange')
plt.axhline(0, color='black', linestyle='--')
plt.title('Year-over-Year % Change')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['imm_rolling'] = df['imm_total'].rolling(window=3).mean()
df['house_rolling'] = df['Housing completions'].rolling(window=3).mean()

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['imm_total'], label='Immigration', alpha=0.5)
plt.plot(df.index, df['house_rolling'], label='3-year Rolling Housing', color='blue', linewidth=2)
plt.plot(df.index, df['imm_rolling'], label='3-year Rolling Immigration', color='red', linewidth=2)
plt.legend()
plt.title('Immigration and Housing with 3-Year Rolling Average')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Correlation Between Variables
We check for linear relationships between immigration and housing completions using a heatmap.

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df[['imm_total', 'Housing completions']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Immigration and Housing')
plt.show()
```

## Annotating Political Periods
Vertical dashed lines mark changes in ruling political parties, helping us spot how transitions may impact trends.

```python
df['party_shift'] = df['Ruling_Party'] != df['Ruling_Party'].shift(1)
change_years = df[df['party_shift']].index

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['imm_total'], label='Immigration')

for year in change_years:
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.7)

plt.legend()
plt.title('Immigration Over Time with Party Changes Marked')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
